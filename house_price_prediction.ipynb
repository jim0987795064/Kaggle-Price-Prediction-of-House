# Data Preprocessing

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

## import dataset

dataset = pd.read_csv("train.csv")
train_dataset = pd.DataFrame(dataset)

dataset = pd.read_csv("test.csv")
test_dataset = pd.DataFrame(dataset)

result = pd.read_csv("sample_submission.csv") 
y_test = pd.DataFrame(result)
y_test = y_test.drop(['Id'], axis=1)



y_train = train_dataset['SalePrice']
x_train = train_dataset.drop(['SalePrice', 'Id'], axis=1)
x_test = test_dataset.drop(['Id'], axis=1)


display(x_test)

## encoode dataset
encoder = LabelEncoder()


for i in x_train.columns:
    if(type(x_train.at[0, i]) == str or x_train[i].hasnans):
        encoded_data = encoder.fit_transform(x_train[i])
        x_train[i] = encoded_data

for i in x_test.columns:
    if(type(x_test.at[0, i]) == str or x_test[i].hasnans):
        encoded_data = encoder.fit_transform(x_test[i])
        x_test[i] = encoded_data



display(x_test)
```

# Decision Tree

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier

price_classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 3, min_samples_split = 6)
price_classifier.fit(x_train, y_train)

predicted_price = price_classifier.predict(x_test)

id = np.arange(1461, 2920)

id = pd.DataFrame(id, columns = ['Id'])
sale_price = pd.DataFrame(predicted_price, columns = ['SalePrice'])
result = pd.concat([id, sale_price], axis=1)

result.to_csv('decisionTree.csv', index=False)
display(result)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV


## model training by GridSearchCV

model = LinearRegression()

grid_params = {'fit_intercept':[True,False],
               'copy_X':[True, False],
               'positive':[True, False]}

grid = GridSearchCV(model, grid_params, cv = 5)
grid.fit(x_train, y_train)
predicted_price = grid.predict(x_test)


## Generate csv

id = np.arange(1461, 2920)

id = pd.DataFrame(id, columns = ['Id'])
sale_price = pd.DataFrame(predicted_price, columns = ['SalePrice'])
result = pd.concat([id, sale_price], axis=1)

result.to_csv('linearRegression.csv', index=False)
display(result)
```
